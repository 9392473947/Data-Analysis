Mobile price classification


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("https://github.com/9392473947/Datasets/raw/main/MobilePriceClassification.csv")
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
plt.figure(figsize=(15,12))
sns.heatmap(df.corr(),annot=True,cmap='cool',linecolor='red',linewidths=1)
```

```python
X=df.iloc[:,:-1].values
```

```python
y=df.iloc[:,-1].values
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X=StandardScaler().fit_transform(X)
```

train test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_test,y_pred)*100
```

```python
print("accuracy score of the logistic regression model is:",accuracy)
```

```python
(unique,counts)=np.unique(y_pred,return_counts=True)
price_range=np.asarray((unique,counts)).T
print(price_range)
```
